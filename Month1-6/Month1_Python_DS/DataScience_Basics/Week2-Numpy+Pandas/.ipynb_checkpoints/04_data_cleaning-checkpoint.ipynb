# 04 Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_csv('sample.csv')
df
```

```python
df.isnull()
df.isnull().sum()
df.info()  # Quick null summary and dtype info
```

```python
df.dropna()  # Drops all rows with any NaN
df.dropna(how='all')  # Drops only rows where all values are NaN
df.dropna(subset=['Name'])  # Drops rows where specific column is NaN
```

```python
df.fillna(value=0)  # Replace NaNs with a static value
df['Age'].fillna(method='ffill')  # Forward fill
df['Name'].fillna(method='bfill')  # Backward fill
df.fillna(df.mean(numeric_only=True))  # Fill with column means
```
