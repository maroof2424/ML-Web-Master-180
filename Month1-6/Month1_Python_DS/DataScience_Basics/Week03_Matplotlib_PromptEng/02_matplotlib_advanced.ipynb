```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.random.rand(6,6)
```

# ***1.   Matplotlib Heatmap***

```python
plt.figure(figsize=(6,5))
plt.imshow(data, cmap="viridis",interpolation="nearest")
plt.colorbar(label="Intensity")
plt.title("Matplotlib Heatmap",fontsize=14)
plt.xticks(range(6), labels=[f"Col {i}" for i in range(1, 7)])
plt.yticks(range(6), labels=[f"Row {i}" for i in range(1, 7)])
plt.grid(False)
plt.show()
```

# ***2. Seaborn Heatmap***





```python
plt.figure(figsize=(6, 5))
sns.heatmap(data, annot=True, fmt=".2f", cmap='coolwarm', cbar_kws={'label': 'Value'})
plt.title("Seaborn Heatmap", fontsize=14)
plt.show()
```
