```python
tips = sns.load_dataset('tips')
sns.set_style("whitegrid")
```

```python
sns.histplot(tips['total_bill'], kde=True, color='purple')
plt.title("Total Bill Distribution")
plt.show()
```

```python
sns.kdeplot(tips['total_bill'], fill=True, color='orange')
plt.title("Total Bill Density Curve")
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=tips, palette='viridis')
plt.title("Average Bill per Day")
plt.show()
```

```python
sns.countplot(x='day', data=tips, palette='coolwarm')
plt.title("Number of Customers per Day")
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='Set2')
plt.title("Bill Distribution by Day")
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='muted')
plt.title("Bill Distribution by Day (Violin)")
plt.show()
```
