```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

## 1. Generate Normal Distribution

```python
mu,sigma = 0,1
```

```python
x = np.linspace(-4,4,1000)
y = norm.pdf(x, mu, sigma)
```

```python
plt.figure(figsize=(8,5))
plt.plot(x,y,label="Î¼=0, Ïƒ=1")
plt.title("Normal Distribution Curve")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()
plt.show()
```

## 2. Properties of Normal Distribution

```python
print("Properties of Normal Distribution:")
print("- Symmetrical about the mean")
print("- Mean = Median = Mode")
print("- 68% data within 1Ïƒ, 95% within 2Ïƒ, 99.7% within 3Ïƒ (Empirical Rule)")
```

## 3. Standard Normal Distribution

### P(-1 <= Z <= 1) ~ 68%

```python
p_between_1 = norm.cdf(1) - norm.cdf(-1)
print("\nP(-1 <= Z <= 1):", round(p_between_1, 3))
```

### P(-2 <= Z <= 2) ~ 95%

```python
p_between_2 = norm.cdf(2) - norm.cdf(-2)
print("P(-2 <= Z <= 2):", round(p_between_2, 3))
```

### P(-3 <= Z <= 3) ~ 99.7%

```python
p_between_3 = norm.cdf(3) - norm.cdf(-3)
print("P(-3 <= Z <= 3):", round(p_between_3, 3))
```

## 4. Applications Example: Heights of People

```python
np.random.seed(42)
heights = np.random.normal(loc=170, scale=10, size=1000)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(heights, kde=True, bins=30, color="skyblue")
plt.title("Normal Distribution Example: Heights of 1000 people")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
print("\nSimulated dataset:")
print("Mean height:", round(np.mean(heights), 2))
print("Std dev of heights:", round(np.std(heights), 2))
```
