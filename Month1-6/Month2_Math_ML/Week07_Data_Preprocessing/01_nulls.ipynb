## Importing Essential Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter("ignore")
```

```python
df = sns.load_dataset("titanic")
df.head()
```

### 1. Drop nulls




```python
df.dropna()
```

### 2. Fill with Mean

```python
df_age_mean = df.copy()
df_age_mean["age"].fillna(df["age"].mean(),inplace= True)
df_age_mean
```

### 3. Forward fill

```python
df_ffill = df.fillna(method="ffill")
df_ffill
```

### 4. Backward fill

```python
df_bfill = df.fillna(method="bfill")
df_bfill
```
